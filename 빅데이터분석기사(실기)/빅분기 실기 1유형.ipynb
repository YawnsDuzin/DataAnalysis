## 3. 이상치 처리 (mpg 데이터)

```python
# 문제11 - clynder가 3인 자동차와 8인 자동차 그룹의 mpg 평균값 차이를 절대값 정수로 출력하시오
df.head()

c_3 = df['mpg'][df['cylinders'] == 3].mean()
c_8 = df['mpg'][df['cylinders'] == 8].mean()

# print(c_3, c_8)

print(int(abs(c_3 - c_8)))
```

```python
# 문제12 - horsepower 변수를 z-score 표준화를 진행하고, 이상치를 구하시오. 이상치는 z값이 1.5를 초과하거나 -1.5미만인 값이다.
#          z-score 표준화 : z = (x - 평균) / 표준편차
```

```python
_mean = df1['horsepower'].mean()
_std = df1['horsepower'].std()
# print(_mean, _std)

df1['z'] = (df1['horsepower'] - _mean) / _std
cond1 = df1['z'] > 1.5
cond2 = df1['z'] < -1.5
print(df1[:][cond1 | cond2].shape[0])
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
df['z'] = ss.fit_transform(df[['horsepower']])

cond1 = df['z'] > 1.5
cond2 = df['z'] < -1.5
print(df[:][cond1 | cond2].shape[0])
```

```python
# 문제13 - mpg컬럼을 min-max정규화를 진행한 후, 0.8보다 큰 값을 가지는 레코드 수를 구하시오.
#          min-max scaling : (x - min) / (max - min)
```

```python
_min = df1['mpg'].min()
_max = df1['mpg'].max()
# print(_min, _max)

df1['min-max'] = (df1['mpg'] - _min) / (_max - _min)

cond1 = df1['min-max'] > 0.8
print(df1[:][cond1].shape[0])
```

```python
from sklearn.preprocessing import MinMaxScaler

mms = MinMaxScaler()

df1['mpg_mm'] = mms.fit_transform(df1[['mpg']])
cond1 =  df1['mpg_mm'] > 0.8
print(df1[:][cond1].shape[0])
```

```python
# 문제14 - weight 컬럼에 대해 상자그림(박스플롯) 기준으로 이상치의 갯수를 구하시오
#          ub = q3 + (1.5 * IQR)
#          lb = q1 - (1.5 * IQR)
#          IQR = q3 - q1

df1 = df.copy()

q3, q1 = df1['weight'].quantile([0.75, 0.25])
IQR = q3 - q1
ub = q3 + (1.5 * IQR)
lb = q1 - (1.5 * IQR)
print(q3, q1, IQR, ub, lb)

cond1 = df1['weight'] < lb
cond2 = df1['weight'] > ub

print(df1['weight'][cond1 & cond2].shape[0])
```

### rank & qcut 함수
- rank 함수 : 순위 매기기
- qcut 함수 : 등급 나누기

```python
# rank 함수
df = sns.load_dataset('car_crashes')
# print(df.head())
# print(df.info())

df1 = df.copy()

# Case1
df1['순위_alcohol'] = df1['alcohol'].rank(method = 'dense', ascending = True)            # 순위 순번 매기기
# df1.sort_values(by='순위_alcohol')
# 등급 나누기
df1['alcohol_grade'] = pd.qcut(df1['순위_alcohol'], q = 5, labels = [1, 2, 3, 4, 5])     # 순위를 등급으로 나누기
df1.sort_values(by = ['순위_alcohol', 'alcohol_grade'])
```
